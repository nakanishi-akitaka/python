# 4章　人間を追い越したコンピューター
## 4-1 機械学習とは何か？
* 人間がプログラミングすると大変な作業を、機械が自動的に学習すること
* 3つの種類がある - 教師あり学習、教師なし学習、強化学習

## 4-2 教師あり学習
* データと正解データのペアをプログラムに与えて、それをもとにルールを学習する方法

## 4-3 教師なし学習
* 正解が分からないデータを分析して、何かしらの構造や法則を見つけるための手法

## 4-4 強化学習
* 報酬を最大化するように学習していく方法

## 4-5 ディープラーニング
* 形式ニューロン  
    複数の入力(1か0)に重み(-0.5、13.2など)をかけて和をとる  
    →　出力(和が正か負かによって、1か0になる)  
    これを組み合わせることでチューリング完全になる  
    組み合わせたものをニューラルネットワークと言う
* パーセプトロン  
    形式ニューロン同士をくっつけたもの　２層  
    入力や出力は1か0だけでなく、実数が扱える
    ニューラルネットワークの一種  
    学習能力を持つ→ブーム  
    線形分離可能なものしか学習できないと判明→ブーム下火  
    バックプロパゲーション発表→再び注目  
* バックプロパゲーション  
    パーセプトロン同士をくっつけたもの　３層  
    線形分離不可能なものでも学習できるようになった  
    層をこれ以上増やしても中々うまくいかない  
* ４層以上でうまくいかない理由  
    多層になると修正のための情報はうまく伝わらなくなる問題 (Vanishing Gradient Problem)  
    ３層あれば大体うまくいってしまう → 最初の層はランダムでも問題ない
* ディープラーニング  
    １層だけで学習→２層を追加＆２層だけで学習→３層～  
    種類　畳み込みニューラルネットワーク(Convolutional neural network)、long short term memoryなど
* 利点  
    特徴抽出を自動的にできる  
    マルチモーダルな特徴（音＋画像　文章＋画像など）が扱いやすくなる  
    過去のアイディアの見直し

# 5章　汎用人工知能を作るには
* 小脳は教師あり学習をしている仮説
    * 小脳 = この状況だったら、こうすればいい、を考える
    * 教師あり学習 = 「状況」が入力、「こうする」が出力
    * 強く支持されている
* 大脳基底核は強化学習をしている仮説
    * 大脳基底核 = いろいろ試した結果を見て、自分にとって得（損）することはもっとする（しない）
    * 教師あり学習と強化学習の違い  
        前者：１つの入力に対して１つの正解を教える  
        後者：色々行動した結果に対して正解か否かを教える  
        例：チェス、1手1手が正解か否かを教えられない  
        最終的な結果＝勝敗は明確  
    * 大脳基底核のドーパミン = 教師あり学習の報酬
    * 有力説
* 大脳新皮質は教師なし学習をしている仮説  
    * 有力視されてない。  
    * 大脳新皮質が何をどう処理してるのか、そもそもよく分からないのが現状。
    * 大脳新皮質 / ディープラーニング　の違い
        * 層間の処理は双方向　/　一方通行  
        * とばすこともある　/　一層づつ進む  
        * 時系列の取り扱いは問題ない　/　苦手  
        * 一度見れば認識できる /　大量のデータが必要  
        * 10層前後　/　膨大(例：152層)  
        * ニューロンの処理速度　数十Hz　/　1GHz  
        * バックプロパゲーションなし　/　あり
