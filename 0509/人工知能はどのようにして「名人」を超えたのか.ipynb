    * 天才(お手本である、～をも超える強さの囲碁)  
    * 人間(を超えるか？肯定派)

> さらには人間そのものからも卒業しようとしています。その物語を、できる限りやさしく語りました。

# はじめに
本書の目的
* なぜ近年、人工知能が急速に進歩したのか？を伝える
* 「知能とは何か？」という問題への、筆者の考えを伝える

# 第1章　将棋の機械学習 - プログラマからの卒業
* コンピュータは「簡単な計算」と「記憶」だけができる  
* いくら、プロ棋士の棋譜を「記憶」しても、それだけでは強くならない
* そもそも知能とは？（著者の理解と解釈）  
    知能 ＝ 探索 ＋ 評価  
* なぜ生物には知能があるのか？  
    知能により未来を予測すると、生存に有利
* 未来を予測するには？
    * エミュレート ＝ 全部探索  
        正確さ・安全が大事な航空・金融・宇宙  
    * シミュレート ＝ 一部探索 ＋ 評価  
        局面が多すぎる将棋・囲碁  
* 評価をプログラムで調整するのが非常に難しい  
* 人間の知恵をコンピュータに与えるのが難しい  
   人間は自分が理解していることをもれなく説明することができない  
   例：なぜその手がいいのか？どのようにして歩いているのか？  
   →　最初期の将棋ソフトは弱い

* 人工知能を人間の知能に近づけるのをやめた → 機械学習  
* チェスと将棋と囲碁の壁＝局面の数ではなく、評価方法  
    * チェスは簡単（駒得で説明できる）
    * 将棋は難しい（言葉で説明できない。味がいい、手厚い、重い）
    * 囲碁はさらに難しい。（何を評価すればいいのか分からない）
* 解決方法 ＝ 機械学習（将棋）　ディープラーニング（囲碁）  
* 教師あり学習＝プロと同じ手を指すように評価方法を自動調整  
* ～の欠点＝教師の数が必要。5万でも足りない。左右対称なもので水増し。  

* プログラマからの卒業  
    プログラマが書いていた、評価にかかわる部分は、機械が自分で学習するようになった
* 本書では、このような人工知能の進歩を、「卒業」と表現する

# 第2章　黒魔術とディープラーニング - 科学からの卒業
* 黒魔術
    * 機械学習で性能を上げるほど、なぜ性能が上がったのかを説明できなくなる  
    * 例：怠惰な並列化  

* ディープラーニング
    * 層を増やすと性能は上がるが学習が難しくなるので、最初はうまくいかなかった
    * 注意深く層を重ねる方法が誕生したことで脚光を浴びる
    * 過学習を防ぐための黒魔術がたくさん。  
        例：ドロップアウト（あえて一部のニューロンを使用不可にして、学習困難にする）
    * 言葉・音声・画像の処理が得意。  
        特に画像。どんなデータでも画像(二次元データ)に直せれば一気に人間を超えられる

* （還元主義的な）科学からの卒業  
    * 還元主義 = 物事を分解し、細部の構造を理解していけば、全体を理解できる  
    * ディープラーニングは、還元主義の考えでは理解できない
    * 知能とは、隠された方程式を解き明かすようなものではなく、モヤモヤとしたよくわからないもの

# 第3章　囲碁と強化学習 - 天才からの卒業
* 人工知能が人間を超えるとき、人間をお手本にするのだろうか？
    * これまでは、守破離の守。どこまでいっても、プロ棋士を超えない。
    * 破・離を達成するのに必要だったのが、強化学習

* 強化学習
    * どんな手を指すか、お手本を渡さない。最終的な結果（勝敗）のみ教えてフィードバックさせる
    * ポナンザでは、合計300コアのマシン数台を何か月も動かして少しずつ強化。膨大な電気代。
    * 強化学習が（現実的な時間で）可能だったのは、教師あり学習と大量のデータがあったからこそ。  
        一から強化学習するのは可能だが、途方もない時間がかかる。  
    * 人工知能の開発においては、必ず大量のデータが必要になる。  
        そのうえで、最初は「教師あり学習」。そしてその後は「強化学習」に移る。

* 囲碁の機械学習が難しい理由＝評価方法そのものが分からない。  
* アルファ碁の３つの武器
    1. モンテカルロ法  
        ランダムな手で最後まで探索した場合の勝率で手を決める。評価を諦めた。
    2. ディープラーニング打ち手予測器  
        囲碁の局面を画像のように扱う  
        →プロっぽい手を計算する
    3. ディープラーニング版評価  
        打ち手予測機を倒す打ち手予測機を機械学習で作成  
        →打ち手予測機同士で対決  
        →勝敗予想器をディープラーニングで作成  
        →勝敗を評価とする。  
* ３つの武器の間での「アンサンブル効果」  
    複数の手法が重ねられることで各段によい結果が得られること。  
    モンテカルロ法とディープラーニングでの評価を組み合わせる。

* 天才からの卒業  
    強化学習＋ディープラーニングによって、人間のお手本よりも上のレベルに到達する

# 第4章　倫理観と人工知能 - 人間からの卒業
* 知性と知能の違い
    * 知性：目的を設計する能力  
    * 知能：目的に向かう道を探す能力  
    * 「知能」は人間を超えたが、「知性」はまだ
* 目的を設計することの重要性
    * PDCA=Plan Do Check Adjustのうち、 人工知能ができるのは探索(Do)と評価(Check)のみ  
        ？調整(Adjust)は機械学習の部分ではなく、自分自身を書き直すプログラム、という意味か
    * 「将棋に勝つ」でなく、「後手玉を詰ます」に目的を変えると、43手詰めを100倍以上の効率で発見
    * 目的を設計するのは、意味と物語にとらわれた、人間ならではの能力（可能性であり限界）
        * 人間は、あらゆることに意味を感じ、物語を読み取ろうとする。それから離れることができない。
        * 人工知能は、意味や物語から自由。目的を設計することはできない。
* 人工知能は目的を設計できるか？
    著者の考え：異なる目的を持つディープラーニング同士をつなぐことで目的を設計できるようになる  
* 人工知能が目的のために手段を選ばないなら、将棋で人間に勝つには人間を絶滅させればいい、となる  
    →倫理観を教える必要性
* 人工知能は人間の倫理観や価値観をも模倣する  
    →人間はいい人であろうとすることになる？  
    * バッグは女性が持つものと考えて翻訳
    * 黒人をゴリラとしてしまう

* 人間からの卒業 = 人間より賢くなる
    その時重要なのは良い人であること？
