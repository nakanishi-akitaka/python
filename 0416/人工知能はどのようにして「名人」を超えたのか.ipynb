# 山本一成『人工知能はどのようにして「名人」を超えたのか？』
山本一成なりの解釈がある点に注意  
すべての章に「～からの卒業」の副題  
* プログラマ(評価を～が教えるのでなく機械学習) 
* 科学(還元主義的～ではない。論理的に説明できない黒魔術で進化)  
* 天才(～＝人間のお手本をも超える強さの囲碁)  
* 人間(を超えるか？肯定派)  

## 第1章
コンピュータは「簡単な計算」と「記憶」だけができる  
知能＝探索＋評価  
エミュレート＝全部探索　　　　正確さ・安全が大事な航空・金融・宇宙  
シミュレート＝一部探索＋評価　局面が多すぎる将棋・囲碁  


人工知能を人間の知能に近づけるのをやめた → 機械学習  
チェスと将棋と囲碁の壁＝局面の数ではなく、評価方法  
* チェスは簡単（駒得で説明できる）
* 将棋は難しい（言葉で説明できない。味がいい、手厚い、重い）
* 囲碁はさらに難しい。（何を評価すればいいのか分からない）
解決方法 ＝ 機械学習（将棋）　ディープラーニング（囲碁）  
教師あり学習＝プロと同じ手を指すように評価方法を自動調整  
～の欠点＝教師の数が必要。5万でも足りない。左右対称なもので水増し。  


## 第2章
機械学習で性能を上げるほど、なぜ性能が上がったのかを説明できなくなる  
＝黒魔術　例：怠惰な並列化  
ディープラーニング＝ニューラルネットワーク  
* 層を増やすと性能は上がるが学習が難しくなるので、最初はうまくいかなかった
* 注意深く層を重ねる方法が誕生したことで脚光を浴びる
* 過学習を防ぐための黒魔術がたくさん。  
    例：ドロップアウト（あえて一部を使用不可にする）
* 言葉・音声・画像の処理が得意。特に画像。画像(二次元データ)に直せれば一気に人間を超えられる


## 第3章
強化学習＝どんな手を指すか、お手本を渡さない。最終的な結果（勝敗）のみ教えてフィードバックさせる  
　ただし、教師あり学習と大量のデータがあったからこそ、強化学習が（現実的な時間で）できた  

囲碁が難しい理由＝評価方法そのものが分からない。  
アルファ碁の３つの武器
1. モンテカルロ法＝ランダムな手で最後まで探索した場合の勝率で手を決める。評価を諦めた。
2. ディープラーニング打ち手予測器＝囲碁の局面を画像のように扱う→プロっぽい手を計算する
3. ディープラーニング版評価　打ち手予測機を倒す打ち手予測機を機械学習で作成→打ち手予測機同士で対決→勝敗予想器をディープラーニングで作成→勝敗を評価とする。  
アンサンブル効果：複数の手法の平均で評価するのがいい。モンテカルロ法とディープラーニングでの評価を組み合わせる。


## 第4章
知性：目的を設計する  
知能：目的に向かって探索する  
PDCA=Plan Do Check Adjustのうち、 人工知能ができるのは探索(Do)と評価(Check)のみ  
？調整(Adjust)は機械学習の部分ではなく、自分自身を書き直すプログラム、という意味か  
著者の考え：ディープラーニングで目的を設計できるようになる  
人工知能は人間の倫理観や価値観をも模倣する  
→人間はいい人であろうとすることになる？  

## おわりに
将棋でもディープラーニングがようやく成功  
合法手が変わることがネックだったが、教えないことで解決
