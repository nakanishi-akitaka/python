# 本：コンピューターで「脳」がつくれるか

# たとえ：ベルトコンベアでの流れ作業で組み立て
* 形式ニューロン  
    複数の入力(0か1)に重みをかけて和をとる  
    →ヘヴィサイドの階段関数→出力(0か1)  
    これを組み合わせることでチューリング完全になる  
    組み合わせたものをニューラルネットワークと言う
* パーセプトロン  
    形式ニューロン同士をくっつけたもの　二層  
    ニューラルネットワークの一種  
    学習能力を持つ→ブーム  
    線形分離可能なものしか学習できないと判明→ブーム下火  
    バックプロパゲーション発表→再び注目  
* バックプロパゲーション  
    パーセプトロン同士をくっつけたもの　三層  
    線形分離不可能なものでも学習できるようになった（？）  
    層をこれ以上増やしても中々うまくいかない  
* うまくいかない理由  
    多層になると修正のための情報はうまく伝わらなくなる問題 (Vanishing Gradient Problem)  
    三層あれば大体うまくいってしまう→最初の層はランダムでも問題ない
* ディープラーニング  
    一層だけで学習→二層を追加＆二層だけで学習→三層～  
    種類　畳み込みニューラルネットワーク(Convolutional neural network)、long short term memoryなど
* 利点  
    特徴抽出を自動的にできる  
    マルチモーダルな特徴（音＋画像　文章＋画像など）が扱いやすくなる  
    過去のアイディアの見直し

# 第5章　汎用人工知能を作るには
* 小脳は教師あり学習をしている仮説
* 大脳基底核は強化学習をしている仮説
* 教師あり学習と強化学習の違い  
    前者：１つの入力に対して１つの正解を教える  
    後者：色々行動した結果に対して正解か否かを教える  
    例：チェス、1手1手が正解か否かを教えられない  
    最終的な結果＝勝敗は明確  
* 大脳新皮質は教師なし学習をしている仮説  
    はあるけれど有力視されてない。  
    そもそも何やってるのかよく分からんのが現状。
* 大脳新皮質＝ディープラーニングではない！ 
    層間の処理は双方向　/　一方通行  
    とばすこともある　/　一層づつ進む  
    時系列の取り扱いは問題ない　/　苦手  
    一度見れば認識できる /　大量のデータが必要  
    10層前後　/　152層  
    ニューロンの処理速度　数十Hz　/　1GHz  
    バックプロパゲーションなし　/　あり
